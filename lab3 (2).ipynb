```python
def assign_ips(device_ids):
    last_num = 2
    
    for i in device_ids:
        ip_address = "192.168.1." + str(last_num)

        print(f"Device ID {i} is assigned IP: {ip_address}")
        
        last_num += 1

D = [501, 502, 503, 504]

assign_ips(D)
```

```python
def grade(mark):
    if mark >= 75:
        return "A"
    elif mark >= 60:
        return "B"
    elif mark >= 40:
        return "C"
    else:
        return "F"

print(f"Mark 85: {grade(85)}")
```

```python
import math
def f(x):
    if x == 0:
        raise ValueError("Domain Error: x cannot be 0")
    return 1 / x
def g(x):
    if x < 0:
        raise ValueError("Domain Error: x must be >= 0")
    return math.sqrt(x)
def h(x):
    if x <= 0:
        raise ValueError("Domain Error: x must be > 0")
    return math.log(x)
def k(x):
    return x**2 + 3*x + 2
def is_in_domain(function, x):
    try:
        function(x)
        return True
    except ValueError:
        return False    
print(is_in_domain(g, 4))   
print(is_in_domain(g, -1)) 
```

```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    c = 0
    for let in s:
        if let in vowels:
            c += 1
    return c
print(count_vowels("HelloWorld"))
```

```python
def encode_pin(x):
    return 2 * x + 3

def decode_pin(y):
    return (y - 3) / 2

original_pin = 1945

encoded_value = encode_pin(original_pin)
print(f"a) Encoded PIN: {encoded_value}")

decoded_value = decode_pin(encoded_value)

is_verified = (decoded_value == original_pin)

print(f"b) Decoded Value: {int(decoded_value)}")
print(f"   Verification: {'Success' if is_verified else 'Failure'}")
```

```python
import math

def p(x):
    return (x**3) - (4 * x) + 1

def q(x):
    if x <= 0:
        raise ValueError("x must be greater than 0")
    # Equation: 1 / sqrt(x)
    return 1 / (x**0.5)  # x**0.5 is the same as square root

def r(x):
    # Domain: x > -1
    if x <= -1:
        raise ValueError("x must be greater than -1")
    # Equation: ln(x + 1)
    return math.log(x + 1)


# --- PART B: The Evaluation Function ---

def evaluate_functions(func_list, x):
    results = []
    
    for f in func_list:
        try:
            # Try to calculate the math
            val = f(x)
            results.append(val)
        except ValueError:
            # If the domain check fails, add the error message instead
            results.append("Domain Error")
            
    return results

# --- TESTING THE CODE ---

# Create a list of the functions we defined above
my_funcs = [p, q, r]

# Let's test with x = -0.5
# p(-0.5) should work
# q(-0.5) should fail (must be > 0)
# r(-0.5) should work (must be > -1)
print("Results for x = -0.5:", evaluate_functions(my_funcs, -0.5))

# Let's test with x = 4
print("Results for x = 4:   ", evaluate_functions(my_funcs, 4))
```
